## <span style="color:blue;"> Programme de l'enseignement *Numérique et Sciences Informatiques* </span>

*Définition* :  
En **informatique**, on conçoit des **algorithmes** (suite d'opérations) qui opèrent sur des **données** (nombres, texte, ...) mais pour faire exécuter ces algorithmes par une **machine** (ordinateur), il faut d'abord les traduire dans un **langage** (python) qui soit compréhensible par la machine.

La NSI s'appuie sur ces **4 concepts clé** de l'informatique :

* Les **données**, qui représentent sous une forme numérique unifiée des **informations** très diverses : textes, images, sons, mesures physiques, sommes d'argent, etc.

* Les **algorithmes**, qui spécifient de façon abstraite et précise des **traitements à effectuer sur les données** à partir d'opérations élémentaires.

* Les **langages**, qui permettent de **traduire les algorithmes abstraits en programmes textuels** ou graphiques de façon qu'ils soient exécutables par les machines.

* Les **machines**, et leurs **systèmes d'exploitation**, qui permettent d'**exécuter des programmes** en enchaînant un grand nombre d'instructions simples, assurent la persistance des données par leur stockage et de gérer les communications. On y inclut les objets connectés et les réseaux.

Le programme comporte **7 thèmatiques** principales :
1. Représentation des données: **types et valeurs de base**
2. Représentation des données: **types construits**
3. Traitement de **données en tables**
4. **Interactions entre l’homme et la machine** sur le Web
5. **Architectures matérielles** et systèmes d’exploitation
6. Langages et **programmation**
7. **Algorithmique**

Pour le programme officiel complet, voir [ici](https://cache.media.eduscol.education.fr/file/SP1-MEN-22-1-2019/26/8/spe633_annexe_1063268.pdf).

## <span style="color:blue;">Les bonnes pratiques de l'informaticien</span>

### La sauvegarde

Toute personne qui travaille avec un ordinateur le sait : **Ca peut planter à tout moment !!**  

Par conséquent, vous devez impérativement **sauvegarder régulièrement vos travaux**, quelque soit le logiciel utilisé !

Vous pouvez sauver votre travail **dans votre répertoire IACA** (dossier *Documents*) mais **pas sur le bureau** car celui-ci est effacé lorsque vous vous déconnectez.

Pour **retrouver votre travail depuis la maison**, le mieux est d'utiliser un **service en ligne**, que ce soit *github*, un simple *cloud* (Pearltrees, Dropbox, GoogleDocs, ...) ou dans votre *espace personnel sur Atrium* (*Tableau de bord* puis *Documents*).

Mais au cas où, vous pouvez aussi utiliser un **support amovible** (clé USB, disque dur, etc). De manière générale, il est toujours plus sûr d'avoir plusieurs sauvegardes différentes.

### Les noms de fichiers

Il est important de donner un **nom qui fasse sens à ses fichiers** (c'est aussi vrai pour les noms de variables d'ailleurs) afin de savoir facilement ce qu'il contient.

Pour éviter les **problème de compatibilité** en fonction des systèmes d'exploitation, **on évitera d'utiliser** :
* Les **lettres accentuées** et autres caractères spéciaux (&, @, ç, %, *, ...).
* Les **espaces** que l'on remplacera par un "_"(*mon_fichier* plutôt que *mon fichier*)
