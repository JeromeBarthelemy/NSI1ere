## 1. Le principe du PageRanking

En 1998, deux jeunes doctorants de l’université de Stanford, Larry Page et Sergey Brin (en collaboration avec Motwani et Wynograd) publiaient un article intitulé "The PageRank Citation Ranking: Bringing Order to the Web" présentant les résultats d’un nouvel algorithme permettant de classer les pages web selon leur popularité et montrant la précision de cet algorithme sur un nouveau moteur de recherche appelé ... Google !

Voilà une version simplifiée de cet algorithme :
- On simule le comportement d’un internaute.
- On démarre au hasard sur une page.
- On suit de façon aléatoire un lien sur cette page.
- Et ainsi de suite...

On impose le fait que dans 15% des cas , l’utilisateur abandonne sa navigation pour repartir d’une page au hasard.

Notre utilisateur va se balader de pages en pages, en faisant son petit tour du net. A chaque fois que l’utilisateur tombe sur une page donnée, cette page gagne un point. Et à la fin, la page ayant le plus de point est alors la page la plus populaire du réseau !

**Remarques sur l'importation de bibliothèques :**
Il y a plusieurs manière d'importer une bibliothèque, selon qu'on veut l'importer en entier ou en partie et selon la syntaxe avec laquelle on souhaite l'utiliser.
- `import random` importe toute la blibliothèque et les fonctions s'utilisent comme des méthodes du module random, par exemple `x=random.random()` pour tirer un nombre aléatoire dans `x`.
- `from random import random` n'importe que la fonction random du module random et cette fonction s'utilise comme si elle avait été directement définie dans le programme en cours, `x=random()` par exemple.
- `from random import *` importe toutes les fonctions du module random et ces fonctions s'utilisent là aussi comme si elles avaient été directement définies dans le programme en cours, `x=random()` par exemple.
