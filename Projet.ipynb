## <span style="color:blue;"> Ce qu'en dit le programme officiel </span>

Un enseignement d’informatique ne saurait se réduire à une présentation de concepts ou de méthodes sans permettre aux élèves de se les approprier en développant des **projets applicatifs**.

Une part de l’horaire de l’enseignement d’au moins **un quart du total en classe de première** doit être réservée à la conception et à l’élaboration de **projets** conduits par des **groupes de deux à quatre élèves**.

Les projets réalisés par les élèves, sous la conduite du professeur, constituent un apprentissage fondamental tant pour la compréhension de l’informatique que pour l’acquisition de compétences.  
En classe de première comme en classe terminale, ils peuvent porter sur des problématiques issues d’autres disciplines et ont essentiellement pour but d’imaginer des solutions répondant à l’expression d’un besoin ; dans la mesure du possible, il convient de **laisser le choix du thème du projet aux élèves eux-mêmes**.

Il peut s’agir d’un approfondissement théorique des concepts étudiés en commun, d’une application à d’autres disciplines telle qu’une simulation d’expérience, d’un travail sur des données socioéconomiques, du développement d’un logiciel de lexicographie, d’un projet autour d’un objet connecté ou d’un robot, de la conception d’une bibliothèque implémentant une structure de données complexe, d’un problème de traitement d’image ou de son, d’une application mobile, par exemple de réalité virtuelle ou augmentée, du développement d’un site Web associé à l’utilisation d’une base de données, de la réalisation d’un interprète d’un mini-langage, d’un programme de jeu de stratégie, etc.

### 2. Des outils pour collaborer en ligne sur un projet

Quand on est dans une même pièce, il est facile de mettre ses idées en commun sur une feuille ou sur un tableau blanc, mais quand on est chacun chez soi, c'est un peu plus compliqué. Heureusement, des outils en ligne existent :

- https://postit.colibris-outilslibres.org/ : Ce site permet de collaborer à plusieurs simultanément sur un tableau blanc sur lequel on place des post-it. Libre, gratuit et facile, c'est idéal pour un petit brainstorm chacun chez soi. Si vous choisissez projet comme nom de votre tableau, vous aurez directement les colonnes "A réaliser", "En cours" et "Terminée" qui seront crées. Mais vous pouvez en créer d'autres et les renommer comme bon vous semble.
- https://www.wopits.com : Ce site fait globalement la même chose que le précédent mais il est un peu plus complet. En contrepartie, il nécessite la création d'un compte (gratuit).
- https://github.com/ : Github, vous connaissez, on l'utilise depuis le début de l'année. Au départ, il n'est pas fait pour héberger des cours mais plutôt pour héberger des projets informatiques. A ce titre, il dispose de nombreuses options de versioning (sauvegarde de l'historique du projet, création de branches, ...) mais aussi d'un outil de gestion de projets(https://www.educba.com/github-projects/).
- https://trello.com/ : Trello, pour finir, est un autre outil collaboratif connu et très complet pour la gestion de projet. Il faut créer un compte (gratuit) mais vous pouvez utiliser des modèles déjà contruit comme par exemple le modèle conduite de projet.
- https://www.notion.so/ : Finalement, un dernier pour la route que je viens de découvrir, il est assez semblable à Trello pour la partie gestion de projets mais il peut aussi faire beaucoup d'autres choses et vous aider dans votre vie de tous les jours d'étudiant. Vous trouverez [ici](https://www.debugbar.com/fr/notion-trello-comparaison/) une comparaison entre Trello et Notion.

Il existe de nombreuses autres solutions de cloud qui permettent aussi de collaborer (GoogleDocs, Atrium, Dropbox, ...) mais elles ne sont pas spécialiement adaptées à la gestion de projets.

## <span style="color:blue;">Evaluation</span>

### Evaluation continue

Tout au long du projet (idéalement chaque semaine), j'essaierai de vous donner quelques petites tâches à accomplir pour vérifier l'état d'avancement de votre projet. Cela donnera lieu à de petites notes régulières (probablement sur 5), qui toutes ensemble formeront une note de suivi de projet.

### Evaluation finale

Au troisième trimestre, le projet devra être terminé et être soutenu à l'oral. Cela aboutira à une note sur 20 (coefficient 3) qui prendra en compte à la fois la qualité de la soutenance orale et la qualité du projet lui-même.
